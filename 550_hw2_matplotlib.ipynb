# Answer1: Plotting graph for oil price vs time from oil_prices.csv 

```python
import csv

import matplotlib.pyplot as plt
#empty lists to store values from the csv file
keys = []
values = []
with open('oil_price.csv', "r") as fh:
    fh = csv.reader(fh, delimiter=',')
    for row in fh:
        keys.append(row[0])
        values.append(row[1])

keys.pop(0)#removing headers
#making a step of 260 to get every year september value to make it even
years = keys[::260]
values.pop(0)#removing headers
prices = values[::260]
plt.figure(figsize=(10,7))
plt.xlabel('Time', fontweight='bold', fontsize=15,color = 'Red')
plt.ylabel('Price', fontweight='bold', fontsize=15,color = 'Red')
plt.title('Oil price variation with time ', fontweight='bold', fontsize=18)


plt.plot(years,prices)
```

# Answer2: PART-1-bar plot showing the number of samples for each body site

# Answer3: PART-1- a plot of position vs the number of base pairs read for 3 different intial position within the genome


```python
import random
import matplotlib.pyplot as plt

genome = ''
fh = open('ecoli_genome.fasta', 'rt')
line = fh.readlines()[1:]
for i in line:
    genome += i
 


class base_walker: 
    def __init__(self):
        pass
    def one_step(self):
        pass
    def steps(self,n=1):
        for i in range(n):
            self.one_step()

class random_walker(base_walker):
    def __init__(self,position):
        self.position = position
        self.walk = [self.position]
    def one_step(self):
        rand =  random.choice(genome)
        if rand == 'A' or rand == 'T':
            self.position += 1
        elif rand == 'C' or rand == 'G':
            self.position -= 1
        self.walk.append(self.position)
    def get_walk(self):
        return self.walk
 


num_walks = 3
total_positions = []
label = []
position = random.random()

for i in range(num_walks): 
    walker = random_walker(position)
    walker.steps(100)
    positions = walker.get_walk()
    total_positions.append(positions)
    label.append('Random genome walk ' + str(i+1)) 

 
plt.figure(figsize=(19,10))
plt.subplot(1, 2, 1)
for walk in total_positions: 
    plt.plot(walk)


plt.ylabel('Position', fontname ='Comic Sans MS',fontweight='bold', fontsize=16)
plt.xlabel('Number of base pairs read', fontname ='Comic Sans MS',fontweight='bold', fontsize=16)
plt.title('E. coli genome Random walk', fontname ='sans-serif',fontweight='bold', fontsize=19)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.legend(labels=label) 
```

# PART-2- a histogram of the random walk displacements over one walk for 200 walks; each consecutive random walk starts where the previos one finishes

```python
import random
import matplotlib.pyplot as plt

genome = ''
fh = open('ecoli_genome.fasta', 'rt')
line = fh.readlines()[1:]
for i in line:
    genome += i
 


class base_walker: 
    '''
    class defined to read number of step taken
    and increement to next step
    
    '''
    def __init__(self):
        pass
    def one_step(self):
        pass
    def steps(self,n=1):
        for i in range(n):
            self.one_step()

class random_walker(base_walker):
    '''
    class defined to calculate random walker
    next steps by reading genome string
    
    '''
    def __init__(self,position=0):
        self.start = position
        self.position = position
        self.walk = [self.position]
    def one_step(self):
        rand =  random.choice(genome)
        if rand == 'A' or rand == 'T':
            self.position += 1
        elif rand == 'C' or rand == 'G':
            self.position -= 1
        self.walk.append(self.position)
    def get_walk(self):
        return self.walk
    
    def get_displacement(self):
        return (self.position - self.start)
 


num_walks = 200
#empty list to store displacement value from each 100 step walk
displacement = []

last_position = 0

for i in range(num_walks): 
    walker = random_walker(last_position)
    walker.steps(100)
    displace= walker.get_displacement()
    displacement.append(displace)
    last_position = walker.position
     

 
plt.figure(figsize=(15,10))
plt.subplot(1, 2, 2)
plt.hist(displacement, color = '#A09F93',label = 'Positions',histtype = 'stepfilled')
plt.legend()




plt.ylabel('Position', fontname ='Comic Sans MS',fontweight='bold', fontsize=16)
plt.xlabel('Number of base pairs read', fontname ='Comic Sans MS',fontweight='bold', fontsize=16)
plt.title('Ecoli random walk displacements over one walk for 200 walks', fontname ='sans-serif',fontweight='bold', fontsize=19)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
```
